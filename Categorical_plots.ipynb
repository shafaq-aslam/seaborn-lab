### Figure level function -> `catplot`

```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Stripplot
- one column should be categorical and one is numerical

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill')
```

```python
# jitter
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = False)
```

```python
# using catplot
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip')
```

```python
# jitter
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip', jitter = 0.1)
```

```python
# hue 
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip', hue = 'sex')
```
