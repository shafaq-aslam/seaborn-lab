# Relational Plots

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot --> Axes level
- lineplot --> Axes level

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

## ScatterPlot

```python
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
## relplot --> figure level --> Square shape --> recommended 
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
# hue 
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex')
```

```python
# style
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue = 'sex',  style = 'time')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex',  style = 'time')
```

```python
# size
# style
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue = 'sex',  style = 'time', size = 'size')
```

```python
# style
sns.relplot(data=tips, x = 'total_bill', y = 'tip',kind = 'scatter', hue = 'sex',  style = 'time', size = 'size')
```

## Lineplot

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

```python
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', kind = 'line')
```

```python
temp_df2 = gap[gap['country'].isin(['Afghanistan', 'Pakistan', 'China', 'India'])]
temp_df2
```

```python
# hue
sns.relplot(kind = 'line', data = temp_df2, x = 'year', y = 'lifeExp', hue = 'country')
```

```python
sns.lineplot(data = temp_df2, x = 'year', y = 'lifeExp', hue = 'country')
```
