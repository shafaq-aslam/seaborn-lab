# Matrix Plot

- Heatmap --> axes level
- Clustermap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
gap = px.data.gapminder()
gap
```

## Heatmap
- Plot rectangular data as a color-encoded matrix

```python
temp_df1 = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
temp_df1
```

```python
sns.heatmap(temp_df1)
```

```python
# to increase size
plt.figure(figsize = (10, 10))
sns.heatmap(temp_df1)
```

```python
# annot
temp = gap[gap['continent'] == 'Europe']
temp
```

```python
temp_df2 = temp.pivot(index = 'country', columns = 'year', values = 'lifeExp')
temp_df2
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(temp_df2)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(temp_df2, annot = True)
```

```python
# linewidth
plt.figure(figsize=(10,10))
sns.heatmap(temp_df2, annot = True, linewidth = 0.5)
```

```python
# cmap
plt.figure(figsize=(10,10))
sns.heatmap(temp_df2, annot = True, linewidth = 0.5, cmap = 'cividis')
```

## Clustermap
- Plot a matrix dataset as a hierarchically-clustered heatmap.
- This function requires scipy to be available.

```python
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
